```python
exercise.head()
```

```python
exercise['Equipment Needed']=exercise['Equipment Needed'].fillna("None")
```

```python
exercise
```

```python
exercise["text_features"] = exercise[["Name of Exercise", "Target Muscle Group", "Difficulty Level", "Equipment Needed"]].astype(str).apply(lambda x: ' '.join(x), axis=1)
```

```python
exercise
```

```python
text_sentences = exercise["text_features"].str.split()
```

```python
text_sentences.head()
```

```python
from gensim.models import Word2Vec
```

```python
word2vec_model=Word2Vec(sentences=text_sentences,vector_size=50,window=5,min_count=1)
```

```python
def get_embedding(text):
    words = text.split()
    valid_words = [word2vec_model.wv[word] for word in words if word in word2vec_model.wv]
    return np.mean(valid_words, axis=0) if valid_words else np.zeros(word2vec_model.vector_size)
```

```python
text_embeddings = np.vstack(exercise["text_features"].apply(get_embedding))
```

```python
text_embeddings[0]
```

```python
#Scaling the data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
num_features
```

```python
# combining both features text and num features
```

```python
combined_features=np.hstack([text_embeddings,num_scaled])
```
