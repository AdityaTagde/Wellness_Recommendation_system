```python
import pandas as pd
import numpy as np
import tensorflow_hub as hub
import tensorflow_text
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
# ✅ Load Pre-trained BERT Model from TensorFlow Hub
bert_model = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

```python
bert_model
```

```python
med_df=pd.read_csv('meditation.csv')
```

```python
med_df
```

```python
med_df['text_features']=med_df['Name']+" "+med_df['Description']
```

```python
med_df
```

```python
embedding=bert_model(med_df['text_features'].tolist()).numpy()
```

```python
embedding
```

```python
with open('meditation_recommender.pkl','wb') as f :
  pickle.dump((med_df,embedding),f)
```
